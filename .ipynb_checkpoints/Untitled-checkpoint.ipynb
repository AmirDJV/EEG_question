```python
#Functions

def makeSound(freq = 6000, # Hz
              duration = 3000): # millisecond
    #import winsound
    winsound.Beep(freq, duration)

def savePythonObject(objectName, fname):
    #import pickle
    outfile = open(fname,'wb')
    pickle.dump(objectName,outfile, protocol=2)
    outfile.close()
    
def readPythonObject(path):
    #import pickle
    infile = open(path,'rb')
    objectName = pickle.load(infile)
    infile.close()
    return objectName

def getSubjecctNumber(file):
    s = os.path.basename(file)[-32:].upper().replace(" ", "")
    start = s.find('N')
    end = start + 4
    s = s[start:end]
    return s
#    
#def getPartNumber(file):
#    s = os.path.basename(file)[0:-8].upper().replace(" ", "")
#    start = s.find('PART')
#    end = start + 5
#    s = s[start:end]
#    return s
#
def getCue(file):
    s = os.path.basename(file)[-32:].upper().replace(" ", "")
    start = s.find('F0')
    end = start + 4
    s = s[start:end]
    return s
```
